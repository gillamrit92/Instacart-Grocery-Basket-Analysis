## Create a histogram of the “order_hour_of_day” column.

```python
hist = instacart_project['order_hour_of_day'].plot.hist(bins = 24)
```

The above histogram represents the times of day when most orders are placed. The most customers are preferred to order the products during mid-days or at the evening that could be between 8am and 4pm.  

## Create a bar chart from the “loyalty_flag” column.

```python
bar = instacart_project['loyalty_flag'].value_counts().plot.bar()
```

## Check whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day.

```python
# Create a random sample of data set by generating a seed
np.random.seed(4)
```

```python
# Create a dev list based on the data split (70/30) distribution
dev = np.random.rand(len(instacart_project)) <= 0.7
```

```python
# Split the dataframe into two samples i.e. big and small
big = instacart_project[dev]
small = instacart_project[~dev]
```

```python
# Check the dataframe length 
len(instacart_project)
```

```python
# check the length of both samples
len(big) + len(small)
```

Both dataframe's lengths are same. 

```python
# Reduce the smaple to columns that are only required for analysis
instacart_2 = small [['order_hour_of_day','prices']]
```

```python
# Check the reduced sample dataset
instacart_2.head(10)
```

```python
# Create a Line chart of reduced sample data
line = sns.lineplot(data = instacart_2, x = 'order_hour_of_day',y = 'prices')
```

## Determine whether there’s a connection between age and family situation by creating a line chart exploring the connections between age and number of dependents.

```python
# Reduce previously made sample to new sample that only need required columns
instacart_3 = small [['age','n_dependants']]
```

```python
# Check the new sample dataset
instacart_3.head(10)
```

```python
# Create a line chart
line_2 = sns.lineplot(data = instacart_3, x = 'age',y = 'n_dependants')
```

The line chart doesn't show any strong relationship between age and no. of dependants. 

## Create a scatterplot that explore whether there’s a connection between age and spending power (income)

```python
# Reduce previously made sample to new sample that only need required columns
instacart_4 = small [['age','income']]
```

```python
# Create a scatterplot
scatterplot = sns.scatterplot(data = instacart_4, x = 'age', y = 'income')
```

### The above scatterplot clearly shows that there is a noticeable upward trend in income starting at the age of 40 as compared to those under 40. This suggests that income tends to increase with age.
