# Merging dataframes

### We have a common key in both data sets is 'product_id' so we can say when merging these two Dataframes using the 'product_id' column, an inner join is the most appropriate choice. 

## Step 1:

```python
# First of all, test the merge without overwriting with merge flag 'match'
pd.merge(orders_product_combined , df_prods , on = ['product_id'] , how = 'inner', indicator = 'match')# 
```

## Step 2: 

```python
# Merge 'orders_product_combined' and 'df_prods'
df_ords_prods_merge = orders_product_combined.merge(df_prods , on = ['product_id'] , indicator = 'match' )
```

```python
# Now, drop the flag column '_merge' from above merge
df_ords_prods_merge.drop('_merge', axis = 1 )
```
